Теперь попробуем метод ближайших соседей 

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=7)
```

```python
neigh.fit(X_train,y_train)
```

Точность на обучающей выборке

Видим, что есть небольшое переобучение. Попробуем подобрать гиперпараметр, чтобы улучшить ситуацию.

```python
K = range(1,60)
scores = []

max_score = -10
k_max_score = 0
for k in K:
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X_train,y_train)
    scr = neigh.score(X_test,y_test)
    if(scr > max_score):
        max_score = scr
        k_max_score = k
    scores.append(scr)
```

```python
plt.plot(K,scores)
plt.xlabel('k')
plt.ylabel('accuracy')
pass
```

```python
print(max_score,k_max_score)
```

```python
neigh = KNeighborsClassifier(n_neighbors=26)
neigh.fit(X_train,y_train)
```

```python
neigh.score(X_train,y_train)
```

```python
neigh.score(X_test,y_test)
```

**Вывод**: модель чему-то обучилась и смогла распространить свои знания на тестовую выборку, то есть переобучения нет. Однако хотелось бы результат лучше, поэтому попробуем использовать ансамблевые классификаторы

```python
from sklearn.ensemble import RandomForestClassifier
N = range(1,60)
scores = []

max_score = -10
n_max_score = 0
for n in N:
    clf = RandomForestClassifier(n_estimators=n)
    clf.fit(X_train,y_train)
    scr = clf.score(X_test,y_test)
    if(scr > max_score):
        max_score = scr
        n_max_score = n
    scores.append(scr)
```

```python
print(max_score,n_max_score)
```

```python
plt.plot(N,scores)
plt.xlabel('n_estimators')
plt.ylabel('accuracy')
pass
```

Мы видим, что ситуация повторилась. Попробуем добавить какие-нибудь нелинейности с помощью нелинейного ядра.

```python
from sklearn.svm import SVC
svc = SVC(gamma='auto')
svc.fit(X_train,y_train)
```

```python
svc.score(X_train,y_train)
```

Эффект такой же. Попробуем понизить размерность до 2 через анализ главных компонент (PCA), чтобы посмотреть, что же творится.

```python
from sklearn.decomposition import PCA
principal=PCA(n_components=2,svd_solver='randomized')
principal.fit(X)
trf = principal.transform(X)
```

```python
trf.shape
```

```python

plt.figure(figsize=(10,10))
plt.scatter(trf[:,0],trf[:,1],c=["r" if val == 0.0 else "g" for val in y],s = 2.0)
plt.xlabel('pc1')
plt.ylabel('pc2')
pass
```

Мы видим, что данные действительно имеют какую-то стуктуру, которую можно было бы использовать.
Возможно, больше всего нам мешает несбалансированность классов. 
Воспользуемся техникой SMOTE (Synthetic Minority Over-sampling Technique), чтобы перебалансировать классы. 

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)
```

```python
X_res_train, X_res_test, y_res_train, y_res_test = train_test_split(X_res, y_res, test_size = 0.25, random_state = 42)
```

```python
K = range(1,60)
scores = []

max_score = -10
k_max_score = 0
for k in K:
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X_res_train,y_res_train)
    scr = neigh.score(X_res_test,y_res_test)
    if(scr > max_score):
        max_score = scr
        k_max_score = k
    scores.append(scr)
```

```python
print(max_score,k_max_score)
```

```python
plt.plot(K,scores)
plt.xlabel('k')
plt.ylabel('accuracy')
pass
```

Точность получилась лучше, но $1$ сосед -- это, конечно, немного странно.

```python
from sklearn.svm import SVC
svc = SVC(gamma='auto')
svc.fit(X_res_train,y_res_train)
```

```python
svc.score(X_res_test,y_res_test)
```

Всё равно результат не очень. Лучше всего справился k-ближйших соседей с $1$ соседом. Матрица ошибок в этом случае:

```python
neigh = KNeighborsClassifier(n_neighbors=1)
neigh.fit(X_res_train,y_res_train)
```

```python
neigh.score(X_res_train,y_res_train)
```
